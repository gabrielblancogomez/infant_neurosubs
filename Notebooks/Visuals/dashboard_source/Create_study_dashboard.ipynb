```python
import pandas as pd
from IPython.display import display, HTML, FileLink
import os
```

```python
# Load demographics data
demographics_df = pd.read_csv('demographics_table_2025_04_23.csv')

# Display first 5 rows to verify
display(demographics_df.head())

# Calculate total sample size
total_n = demographics_df.loc[(demographics_df.Characteristic == 'Male') | (demographics_df.Characteristic == 'Female') , 'n'].sum()

print(f"Total participants: {total_n}")
```

```python
def generate_html_table(df):
    """Converts demographics DataFrame to styled HTML table"""
    html = ['<table class="demographics-table">']
    
    # Table header
    html.append("""
    <thead>
        <tr>
            <th>Characteristic</th>
            <th>n</th>
            <th>%</th>
        </tr>
    </thead>
    <tbody>""")
    
    # Table body
    for _, row in df.iterrows():
        if pd.isna(row['n']) or row['n'] == '':
            # Category header row
            html.append(f'<tr class="category-row"><td colspan="3">{row["Characteristic"]}</td></tr>')
        else:
            # Data row
            html.append(f"""
            <tr>
                <td>{row["Characteristic"]}</td>
                <td>{int(row["n"])}</td>
                <td>{row["%"]}</td>
            </tr>""")
    
    html.append("</tbody></table>")
    return '\n'.join(html)

```

```python
# Read current HTML file
with open('dashboard3.html', 'r', encoding='utf-8') as f:
    html_content = f.read()

# Replace placeholders
updated_html = html_content.replace(
    '<!-- DEMOGRAPHICS-TABLE-PLACEHOLDER -->',
    generate_html_table(demographics_df)
).replace(
    '<span id="total-n">144</span>',
    f'<span id="total-n">{total_n}</span>'
)

# Save updated file (creates new version)
output_filename = f'dashboardV4.html'
with open(output_filename, 'w', encoding='utf-8') as f:
    f.write(updated_html)

# Create download link
display(FileLink(output_filename))
print("HTML file updated successfully!")
```

```python
 #Read current HTML
with open('dashboard3.html', 'r', encoding='utf-8') as f:
    html_content = f.read()

# Generate new table HTML
new_table = generate_html_table(demographics_df)

# Perform the replacement - METHOD 1 (Recommended)
updated_html = html_content.replace(
    '<!-- DEMOGRAPHICS-TABLE-PLACEHOLDER -->', 
    new_table
)

# Or METHOD 2 (Alternative - if placeholder has extra spaces)
updated_html = html_content.replace(
    '''<div id="demographics-table-container">
        <!-- DEMOGRAPHICS-TABLE-PLACEHOLDER -->
    </div>''',
    f'''<div id="demographics-table-container">
        {new_table}
    </div>'''
)

# Save the file
with open('dashboard_V4.html', 'w', encoding='utf-8') as f:
    f.write(updated_html)
```
